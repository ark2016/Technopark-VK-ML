```python
test = pd.read_pickle("test.pkl")
test
```

```python
train["title"] = train["title"].apply(lambda x: translator.translate(x, dest='en').text)
```

```python
train_t = []
```

```python
import time
```

```python
time.sleep(1) 
```

```python
for j in range(0,len(train["title"]),1000):
    for i in train["title"].loc[j : min(j + 1000, len(train["title"]))]:
        train_t.append(translator.translate(i, dest='en').text)
    time.sleep(1) 
train_t
```

```python
test["title"] = test["title"].apply(lambda x: translator.translate(x, dest='en').text)
```
